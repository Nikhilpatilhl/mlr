```python
user_data=dt1.pivot_table(index='UD',columns='BT',values='BR')
```

```python
user_data.fillna(0,inplace=True)
user_data
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
sim = 1-pairwise_distances(user_data.values,metric='cosine')
sim
```

```python
sim_data=pd.DataFrame(sim)
```

```python
sim_data.shape
```

```python
sim_data.index= user_data.index
sim_data.columns = user_data.index
```

```python
def get_reco(UD):
    tem = list(sim_data.sort_values([UD],ascending=False).head(100).index)
    book_list = list(dt1[dt1["UD"].isin(tem)]["BT"])
    #print(len(book_list)
    return set(book_list)-set(dt1[dt1["UD"]==UD]["BT"])
```
